```python
#OOP of Matplotlib
a=np.linspace(0,10,11)
b=a**4
```

```python
b
```

```python
x=np.arange(0,10)
y=2*x
```

```python
y
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b);
```

```python
type(axes)
```

```python
axes.shape
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(x,y)
fig.tight_layout()
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b);
fig.subplots_adjust(wspace=1,hspace=0.5)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=200)
axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y LABEL 1,0')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(x,y);
axes[1][1].set_title('TITLE 1,1')

fig.suptitle("Figure Level",fontsize=16)

plt.tight_layout()
fig.savefig('new_subplots.png',bbox_inches='tight')
```

# Lecture 53- Matplotlib Styling-Legends

```python
x=np.linspace(0,10,11)
x
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,x,label='X vs X');
ax.plot(x,x**2,label="X vs X^2")
ax.legend(loc=(-0.5,0.5));
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x,x,label='X vs X');
ax.plot(x,x**2,label="X vs X^2")
ax.legend(loc=(-0.5,0.5));
```

# 54.Matplotlib Styling

```python
x=np.linspace(0,10,11)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# ax.plot(x,x)
ax.plot(x,x,color='#d1b873',label='x vs x',lw=10,ls=':') #RGB HEX Code
ax.plot(x,x+1,color='#65c1e6',label="x vs x+1",linewidth=2,linestyle='-.');
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# ax.plot(x,x)
# ax.plot(x,x,color='#d1b873',label='x vs x',lw=10,ls=':') #RGB HEX Code
lines=ax.plot(x,x+1,color='purple',linewidth=5);
# lines[0].set_dashes([5,2,5,2])
lines[0].set_dashes([1,2,1,2,10,2,1,2,1,10])
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# ax.plot(x,x)
# ax.plot(x,x,color='#d1b873',label='x vs x',lw=10,ls=':') #RGB HEX Code
ax.plot(x,x+1,color='purple',linewidth=5,marker='s',ls='--',ms=15,
       markerfacecolor='red',markeredgewidth=10,markeredgecolor='orange');
ax.plot(x,x,color='red',linewidth=2,marker='+',ls='--',markersize=20)
# lines[0].set_dashes([5,2,5,2])
# lines[0].set_dashes([1,2,1,2,10,2,1,2,1,10])
```
